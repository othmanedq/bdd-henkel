```python
import pandas as pd

file_path_leroy = 'DATA Leroy Merlin.xlsx'
file_path_bricomarche = 'DATA Bricomarché.xlsx'

leroy_data = pd.read_excel(file_path_leroy, sheet_name=None)
bricomarche_data = pd.read_excel(file_path_bricomarche, sheet_name=None)
```

```python
leroy_data.keys()
```

```python
leroy_total_visits_2022 = leroy_data['VISITES ANNUELLES 22 23']['Total visites annuelles\n2022'].sum()
leroy_total_visits_2023 = leroy_data['VISITES ANNUELLES 22 23']['Total visites annuelles\n2023'].sum()

leroy_total_visits_2022, leroy_total_visits_2023
```

```python
leroy_sales_data = leroy_data['CA SELL IN LM 22 23']

leroy_sales_data.head()
```

```python
monthly_columns_2023 = [
    'Janvier\n2023', 'Février\n2023', 'Mars\n2023', 'Avril\n2023', 
    'Mai\n2023', 'Juin\n2023', 'Juillet\n2023', 'Août\n2023', 
    'Septembre\n2023', 'Octobre\n2023', 'Novembre\n2023', 'Décembre\n2023'
]

monthly_sales_2023 = leroy_sales_data[monthly_columns_2023].sum()

highest_sales_month_2023 = monthly_sales_2023.idxmax()
highest_sales_value_2023 = monthly_sales_2023.max()

highest_sales_month_2023, highest_sales_value_2023
```

```python

annual_sales_2022 = leroy_sales_data.groupby('Famille')['Total\nannuel\n2022'].sum()
annual_sales_2023 = leroy_sales_data.groupby('Famille')['Total\nannuel\n2023'].sum()

sales_trends = pd.DataFrame({
    'Total 2022': annual_sales_2022,
    'Total 2023': annual_sales_2023
})

sales_trends['Change (%)'] = ((sales_trends['Total 2023'] - sales_trends['Total 2022']) / sales_trends['Total 2022']) * 100

import ace_tools_open as tools; tools.display_dataframe_to_user(name="Sales Trends 2022 to 2023", dataframe=sales_trends)
```

```python
bricomarche_sales_data = bricomarche_data['Données brutes SELL IN 22 23']

bricomarche_sales_data.head()
```

```python
annual_sales_bricomarche_2022 = bricomarche_sales_data.groupby('Famille')['Total\nannuel\n2022'].sum()
annual_sales_bricomarche_2023 = bricomarche_sales_data.groupby('Famille')['Total\nannuel\n2023'].sum()

bricomarche_sales_trends = pd.DataFrame({
    'Total 2022': annual_sales_bricomarche_2022,
    'Total 2023': annual_sales_bricomarche_2023
})

bricomarche_sales_trends['Change (%)'] = ((bricomarche_sales_trends['Total 2023'] - bricomarche_sales_trends['Total 2022']) / bricomarche_sales_trends['Total 2022']) * 100

tools.display_dataframe_to_user(name="Bricomarché Sales Trends 2022 to 2023", dataframe=bricomarche_sales_trends)
```

```python


bricomarche_visits_2023 = bricomarche_data['VISITES ANNUELLES 22 23'][['CODE INTERNE UNIQUE DU POINT DE VENTE', 'Total visites annuelles\n2023']]
bricomarche_sales_2023 = bricomarche_sales_data[['Code du Point de Vente', 'Total\nannuel\n2023']]

bricomarche_visits_2023.rename(columns={'CODE INTERNE UNIQUE DU POINT DE VENTE': 'Code du Point de Vente'}, inplace=True)

merged_data_2023 = pd.merge(bricomarche_visits_2023, bricomarche_sales_2023, on='Code du Point de Vente', how='inner')

correlation = merged_data_2023['Total visites annuelles\n2023'].corr(merged_data_2023['Total\nannuel\n2023'])

correlation
```
